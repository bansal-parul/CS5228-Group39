```python
train_data = pd.read_csv("/content/drive/MyDrive/CS5228_Project/Datasets/train_pdf_stock.csv")
test_data = pd.read_csv("/content/drive/MyDrive/CS5228_Project/Datasets/test_pdf_stock.csv")
```

```python
#drop the columns: town, latitude, longitude, block, street_name, lease_commence_date
columns_drop = ["town", "latitude","longitude","block","street_name","lease_commence_date"]
train = train_data.drop(columns_drop, axis =1)
test = test_data.drop(columns_drop, axis =1)
```

```python
train[["year", "month"]] = train["rent_approval_date"].str.split("-", expand = True)
test[["year", "month"]] = test["rent_approval_date"].str.split("-", expand = True)
train[["year", "month"]] = train[["year", "month"]].astype(int)
test[["year", "month"]] = test[["year", "month"]].astype(int)
train = train.drop("rent_approval_date", axis =1 )
test = test.drop("rent_approval_date", axis =1 )
```

```python
train.flat_type.unique()
```

```python
train["flat_type"] = np.where(train["flat_type"] == '2_room', 2,
                              np.where(train["flat_type"] =='3_room',3,
                              np.where(train["flat_type"] == '4_room',4,
                              np.where(train["flat_type"] == '5_room',5,6))))


test["flat_type"] = np.where(test["flat_type"] == '2_room', 2,
                              np.where(test["flat_type"] =='3_room',3,
                              np.where(test["flat_type"] == '4_room',4,
                              np.where(test["flat_type"] == '5_room',5,6))))
```

```python
dummy_region_train = pd.get_dummies(train[["region","flat_model"]], drop_first = True)
dummy_region_test = pd.get_dummies(test[["region","flat_model"]], drop_first = True)
```

```python
train = pd.concat([train,dummy_region_train ], axis =1)
test = pd.concat([test,dummy_region_test ], axis =1)
```

```python
train = train.drop(["flat_model","region"], axis =1)
test = test.drop(["flat_model", "region"], axis =1 )
```

```python
#target encoding

grouped_subzone = train.groupby("subzone").monthly_rent.mean().reset_index()
grouped_subzone.columns = ["subzone", "subzone_encoded"]
train = train.merge(grouped_subzone, on = "subzone", how = "left")
```

```python
test = test.merge(grouped_subzone, on = "subzone", how = "left")
```

```python
grouped_planning_area = train.groupby("planning_area").monthly_rent.mean().reset_index()
grouped_planning_area.columns = ["planning_area", "planning_area_encoded"]

train = train.merge(grouped_planning_area, on = "planning_area", how = "left")
test = test.merge(grouped_planning_area, on = "planning_area", how = "left")

train = train.drop(["planning_area"], axis =1)
test = test.drop(["subzone"], axis =1)
test = test.drop(["planning_area"], axis =1)
train = train.drop(["subzone"], axis =1)
```
